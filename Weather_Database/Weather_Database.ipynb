```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
lats=np.random.uniform(-90.000,90.000,size=2000)
lngs=np.random.uniform(-180.000,180.000,size=2000)
lats_lngs=zip(lats,lngs)
lats_lngs
```

```python
coordinates=list(lats_lngs)
coordinates
```

```python
from citipy import citipy
```

```python
cities=[]
for coordinate in coordinates:
    city=citipy.nearest_city(coordinate[0],coordinate[1]).city_name
    
    if city not in cities:
        cities.append(city)
        
cities        
```

```python
import requests
```

```python
from config import weather_api_key
```

```python
url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + weather_api_key
```

```python
record_count=1
set_count=1
city_data=[]
```

```python
city_url=url+"&q="+"London"
  
    
ci = requests.get(city_url).json()
ci
```

```python
for i,city in enumerate(cities):
    
    if(i%50==0 & i>=50):
      
        set_count+=1
        record_count=1
    
    city_url=url+"&q="+city.replace(" ","+")
    print(city,"cityyyyyyy")
    try:
        city_weather = requests.get(city_url).json()
        city_lat=city_weather["coord"]["lat"]
        city_lng=city_weather["coord"]["lon"]
        city_max_temp=city_weather["main"]["temp_max"]
        city_humidity = city_weather["main"]["humidity"]
        city_clouds = city_weather["clouds"]["all"]
        city_wind = city_weather["wind"]["speed"]
        weather_description=city_weather["weather"][0]["description"]
        country=city_weather["sys"]["country"]
        
        city_data.append({
            "city":city.title(),
             "lat":city_lat,
             "lng":city_lng,
             "max temp":city_max_temp,
             "Humidity":city_humidity,
             "Cloudiness":city_clouds,
             "wind speed":city_wind,
             "Weather description":weather_description,
             "country":country
            })
        
    except KeyError:
        
        print("There is an error for the "+city)
        continue
        
    record_count+=1
```

```python
weather_df=pd.DataFrame(city_data)
weather_df
```

```python
output_data_file = "WeatherPy_Database.csv"
weather_df.to_csv(output_data_file, index_label="City_ID")
```
