```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
lats=np.random.uniform(-90.000,90.000,size=1500)
long=np.random.uniform(-180.000,180.000,size=1500)
lats_long=zip(lats,long)
```

```python
coordinates=list(lats_long)
coordinates
```

```python
from citipy import citipy
```

```python
for coordinate in coordinates:
   print( citipy.nearest_city(coordinate[0],coordinate[1]).city_name,
   citipy.nearest_city(coordinate[0],coordinate[1]).country_code)
```

```python
cities=[]
for coordinate in coordinates:
    city=citipy.nearest_city(coordinate[0],coordinate[1]).city_name
    
    if city not in cities:
        cities.append(city)
        
len(cities)        
```

```python
import requests
```

```python
from config import weather_api_key
```

```python
url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + weather_api_key
print(url)
```

```python
from datetime import datetime
#date=boston_data["dt"]
#datetime.utcfromtimestamp(date).strftime("%m-%d_%y %H:%M:%S")
```

```python
city_data=[]
```

```python
print("Beginning Data Retrieval     ")
print("-----------------------------")
```

```python
record_count=1
set_count=1
```

```python
for i,city in enumerate(cities):
    
    if(i%50==0 & i>=50):
        set_count+=1
        record_count=1
    
    city_url=url+"&q="+city.replace(" ","+")
    
    print(f"record count is :{record_count} and set count is:{set_count} for {city}")
    try:
        city_weather = requests.get(city_url).json()
    
        city_lat = city_weather["coord"]["lat"]
        city_lng = city_weather["coord"]["lon"]
        city_max_temp = city_weather["main"]["temp_max"]
        city_humidity = city_weather["main"]["humidity"]
        city_clouds = city_weather["clouds"]["all"]
        city_wind = city_weather["wind"]["speed"]
        city_country = city_weather["sys"]["country"]  
        print("yyyyy",city.title())
        city_data.append({
            "city":city.title(),
            "lng": city_lng,
            "lat":city_lat,
            "max_temp":city_max_temp,
            "Humidity": city_humidity,
            "Cloudiness": city_clouds,
            "Wind Speed": city_wind,
            "Country": city_country,
            
             
        })
    except KeyError:
        print("KeyError recivied for"+city)
        continue
    print(city_lat,city_lng,city_max_temp, city_humidity,city_clouds,city_wind,city_country)
    record_count+=1
    
```

```python
print("-----------------------------")
print("Data Retrieval Complete      ")
print("-----------------------------")
```

```python
from scipy.stats import linregress
```

```python
def plot_linear_regression(x_values,y_values,title,y_label,text_coordinates):
    (slope,intercept,r_value,p_value,srd_err)=linregress(x_values,y_values)
    print("r_value",r_value)
    regress_values=x_values*slope+intercept
    
    line_eq="y = "+str(round(slope,2))+" x + "+ str(round(intercept,2))
    plt.scatter(x_values,y_values)
    plt.plot(x_values,regress_values,"r")
    plt.annotate(line_eq,text_coordinates, fontsize=15, color="red")
    plt.xlabel("Lattitude")
    plt.ylabel(y_label)
    plt.show()
    
    
```
